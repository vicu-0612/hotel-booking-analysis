```python
import pandas as pd 
import matplotlib.pyplot as plt   # Python plot, math plot library

import seaborn as sns

# Set theme, palette, style and context for figure 
sns.set()
```

```python
# Change family font to serif
plt.rcParams['font.family'] = "serif"
```

#### Line chart

```python
dates = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
study_hours_1 = [6, 7, 5, 8, 6, 2, 0]
study_hours_2 = [8, 7, 4, 7, 5, 5, 3]
```

```python
plt.figure(figsize= (16,5))

plt.plot(dates, study_hours_1, color= "#E98580", marker= "o", linestyle='dashed', label= "Huy")
plt.plot(dates, study_hours_2, color= "#FF7600", marker= "o", linestyle='dashed', label= "Dung")

plt.title(label= "Daily Study Hours", fontsize= 20, loc= "left", pad= 25)
plt.xlabel(xlabel= "Weekdays", fontsize= 13, labelpad= 10)
plt.ylabel(ylabel= "Hours", fontsize= 13, labelpad=10)

plt.legend()

plt.show()
```

#### Bar chart

```python
language_data = pd.read_csv("datasets/language.csv")
language_data.head()
```

```python
# Get top seven languages from data
sorted_lang_data = language_data.sort_values(by= "num_of_users", ascending=False)
lang_col = sorted_lang_data["language"][:7]
user_col = sorted_lang_data["num_of_users"][:7]
```

```python
user_col
```

```python
plt.figure(figsize= (16,5))
plt.bar(x= lang_col, height= user_col, color= "#FFBCBC")
plt.title("Top Seven Programming Languages", fontsize= 20, loc= "left", pad= 30)
# plt.xticks(rotation= 45)
plt.ylabel("Number of Users", labelpad= 10)
plt.show()
```

#### Barh chart

```python
# Get top seven languages from data
sorted_lang_data = language_data.sort_values(by= "num_of_users", ascending=False)
top_seven_data =sorted_lang_data.iloc[:7].sort_values(by="num_of_users")
lang_col = top_seven_data["language"]
user_col = top_seven_data["num_of_users"]
```

```python
plt.figure(figsize= (16,5))
lang_bar = plt.barh(y= lang_col, width= user_col, color= "#FFBCBC")
plt.title("Top Seven Programming Languages", fontsize= 20, loc= "left", pad= 30)
plt.xlabel("Number of Users", labelpad= 10)


for i in range(len(lang_bar)):
    
    info = lang_bar.patches[i].get_bbox().get_points()  # (xmin, ymin), (xmax, ymax)
    x0 = info[0][0]
    y0 = info[0][1]
    x1 = info[1][0]
    y1 = info[1][1]

    text_x = x1 - x0
    text_y = (y1 + y0) / 2

    plt.text(x= text_x + 2000, y= text_y, s= round(text_x), ha= "center")

plt.show()
```

#### Pie chart

```python
ms_data = pd.read_csv("datasets/market-share.csv")
ms_data
```

```python
plt.figure(figsize= (7,7))
plt.pie(x= ms_data['frequency'], 
        explode=[0, 0, 0.05], 
#         labels= ms_data["branch"], 
        colors= ["#DF5E5E", "#E98580", "#FDD2BF"],
        autopct= "%.2f%%",
        shadow= True,
        startangle=90,
        radius= 1,
        counterclock= False,
)
plt.title("Market Share", fontsize= 20, pad= 20)
plt.legend(labels= ms_data["branch"])
plt.show()
```

#### Scatter chart

```python
sat_data = pd.read_csv("datasets/sat-score.csv")
sat_data.head()
```

```python
sat_data.info()
```

```python
sat_data.describe()
```

```python
plt.figure(figsize= (16,5))
plt.scatter(x= sat_data["Reading"], y= sat_data["Writing"])
plt.title("SAT Scores", fontsize= 20, pad= 25, loc= "left")
plt.xlim(0, 800)
plt.ylim(0, 800)
plt.xlabel("Reading")
plt.ylabel("Writing")

plt.show()
```

```python
sat_data.head()
```

```python
us_data = sat_data[sat_data["Nation"] == "US"]
vn_data = sat_data[sat_data["Nation"] == "VN"]
```

```python
plt.figure(figsize= (16,5))

plt.scatter(x= us_data["Reading"], y= us_data["Writing"], label= "American")
plt.scatter(x= vn_data["Reading"], y= vn_data["Writing"], label= "Vietnamese")

plt.title("SAT Scores", fontsize= 20, pad= 25, loc= "left")
plt.xlim(0, 800)
plt.ylim(0, 800)
plt.xlabel("Reading")
plt.ylabel("Writing")

plt.legend(loc= "lower left")

plt.show()
```

#### Time Series Plot

```python
stock_data = pd.read_csv("datasets/stock-data.csv")
close_data = stock_data[["Date", "Close"]]
```

```python
close_data
```

```python
close_data.info()
```

```python
close_data["Date"] = pd.to_datetime(close_data["Date"])
```

```python
close_data.dtypes
```

```python
fig = plt.figure(figsize=(16,5))
plt.plot_date(x= close_data["Date"], y= close_data["Close"], fmt= "-o", color= "#E98580")
plt.xticks(rotation= 45)
plt.title("Stock Market", fontsize= 20, pad= 25, loc= "left")
plt.ylim(5000, 10000)
plt.xticks(ticks= close_data["Date"])
plt.ylabel("Closing Price")
plt.show()
```
